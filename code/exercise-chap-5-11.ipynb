```python
import pandas as pd
import altair as alt
```

```python

pm25_2011_durham = pd.read_csv('https://raw.githubusercontent.com/kirenz/datasets/master/pm25_2011_durham.csv')
```

```python
chart = alt.Chart(pm25_2011_durham).mark_bar().encode(
    x=alt.X("daily_aqi_value:Q", bin=alt.Bin(step=5)),
    y='count()',
   # y=alt.Y("count(daily_aqi_value:Q)/sum(count(daily_aqi_value:Q))", axis=alt.Axis(format='.2f')),
)

text = alt.Chart(pm25_2011_durham).mark_text(dx=2, dy=+10, color='white').encode(
    x=alt.X("daily_aqi_value:Q", bin=alt.Bin(step=5)),
    y='count()',
    text=alt.Text("count(daily_aqi_value):Q/sum(count(daily_aqi_value)):Q", format='.2f')
)

chart.properties(
    width=500,
    height=300
)

chart + text
```
